```python
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
import warnings
warnings.filterwarnings('ignore')

# Remove pandas restrictions:---------------------------------
pd.set_option('Display.max_rows', None)
pd.set_option('Display.max_columns', None)

# Data and pre-processing:-------------------------------------
url="https://raw.githubusercontent.com/erinijapranckeviciene/MLdata/main/Kaggle-data-suic.csv"
data = pd.read_csv(url)
isuic=data.drop(['suicides/100k pop', 'country-year', 'HDI for year', ' gdp_for_year ($) ', 'generation'], axis=1)
```

```python
# Combined regression------------------------------------------
countries = list(set(isuic['country']))
intercepts = []
coefficients = []
for country in countries:
    X = isuic[(isuic['country'] == country)][(isuic['sex'] == "male")]
    X1 = X[["year"]].copy()
    X2 = X[["year","age", "population", 'gdp_per_capita ($)']].copy()
    X2_dummies = pd.get_dummies(X2)
    Y = X['suicides_no']
    regr = linear_model.LinearRegression()
    regr.fit(X2_dummies, Y)
    intercepts.append(regr.intercept_)
    coefficients.append(list(regr.coef_))

Reg = pd.DataFrame(data = {'country': countries, 'intercept': intercepts, 'coefficients': coefficients})
Reg
```
